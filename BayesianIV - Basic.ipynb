# Bayesian Treatment and Response Modeling

The basic idea in this worksheet (and in subsequent ones) is to develop a means of modelling unobserved heterogeneity and selectivity effects in the usual treatment and response model. As a by product, one can recover treatment effects for the individuals in the population, because these treatment effects are treated as parameters in a Bayesian approach, and are therefore estimated along with everything else in the model.

### Latent variables

The method that I will rely on centers on _latent variables_. In a Bayesian setting, these variables can usually be treated as parameters and can be drawn from a conditional distribution like everything else. The best way to illustrate this is to simply dive in and see how it works! 

### A simple linear model

Consider a simple linear model with a response equation:

$$
y_i = X_i \beta + \delta z_i + \epsilon_{yi}
$$

and a treatment equation:

$$
z_i = W_i \eta + \epsilon_{zi}
$$

The second equation explains how the intensity of treatment depends upon (observable) individual characteristics $W_i$ and unobservables $\epsilon_z$, while the first equation describes how outcomes depend upon treatment, $z_i$, observed characteristics $X_i$, and unobserved characteristics $\epsilon_y$. 

One hopes that $Cov(\epsilon_y,\epsilon_z)=0$, so that inferences about the conditional impact of $z$ on $y$ is valid. However, there are often reasons to doubt this. A very big reason is that selection into treatment depends in part on unobservables. For example, you may observe a population of injured people - say football players who have knee injuries. Only some of these players receive surgery. You may not know, or be able to measure only imperfectly, how severe the injuries are, so severity is unobserved. But the more severe injuries are more likely to receive surgery. If you analyzed this data naively, you might be likely to find that $y_i$ - some measure as to how well the player recovered from the injury - is _negatively_ correlated.

### Bayesian methods and latent variables

One of the great things about Bayesian methods is that one can quantify beliefs about how seriously one believes the conditional independence assumption through use of a prior distribution.

Anyways, one is typically interested in some set of parameters in a model and how they are influenced by the data. In the above model, the parameters are $\theta = \beta,\delta,\eta,\sigma_y,\sigma_z$, which are related to the model data $D = y,X,W$ through some distribution(s). Classical estimation proceeds through maximization of a function of the data and parameters. For example, maximum likelihood uses probability theory to create a likelihood function:

$$
L = f(D|\theta)
$$

which is then maximized w. r. t. $\theta$. The idea is to find the parameters that maximize the likelihood of observing the data given the model. Once these maximizing parameters are obtained, some calculus and probability theory is used to derive the distribution of the parameters at the maximum - standard errors, confidence intervals, etc. 

Bayesian methods don't typically involve maximization, and instead start with the thought that if we had a model, and some data for the model, we could then recover the distribution of the parameters and other things we might be interested in directly. As it turns out, the distribution of the parameters is usually implied by the likelihood function. What is the distribution for the parameters? Bayes's rule says that:

$$
f(D|\theta)f(\theta) = f(\theta|D)f(D)
$$

Or

$$
f(\theta|D) = \frac{f(D|\theta)f(\theta)}{f(D)}
$$

Because the unconditional distribution of the data $f(D)$, does not depend upon parameters, this last thing is usually written as:

$$
f(\theta|D) \propto f(D|\theta)f(\theta)
$$

Two things:

1. If we can characterize the parameter distribution directly, why go through the often difficult process of maximizing likelihood, and why not just use the parameter distribution? 

2. Note the presence of $f(\theta)$. This is commonly called a prior distribution of the parameters. In the Treatment-response case, for example, we might wish to somehow characterize our prior belief that the covariance in errors is zero, which would be similar to the usual instrumental variables assumption - that instruments are uncorrelated with errors. 

A further note: [A great paper by Chernozhukov and Hong (2003)](http://papers.ssrn.com/sol3/papers.cfm?abstract_id=420371) shows how one can rely on 1., coupled with MCMC, to estimate otherwise difficult models. [An equally great (!) paper by myself (Baker, 2014)](http://www.stata-journal.com/article.html?article=st0354) discusses some practical examples of stuff like this. 

## Practicalities

As a practical matter, one often does a couple of things to actually implement a Bayesian method, but at the heart of most Bayesian methods is some sort of Markov-chain Monte Carlo (Gibbs, Metropolis-Hastings being common variants) sampling routine. These allow sampling from conditional distributions, which are often easy to work with. There are many proofs out there of convergence. But at this point, suffice it to say that I'm going to describe one of these that uses _latent variables_ for the IV problem. A good place to start with this is by first fleshing out our model, and then looking at the implied likelihood. 

### Latent variables and cross-equation correlation

To begin with, let's first add in an unobserved variable that captures the correlation across our equations. To do this, we suppose that:

$$
\epsilon_{yi} = \pi_y u_i + e_{yi}
$$
and
$$
\epsilon_{zi} = \pi_z u_i + e_{zi}
$$

Here, the latent variable is $u_i$, which is something that we don't see about the individual but is present in (the error terms) of both equations. We assume that $E[u_i] = E[e_{yi}] = E[e_{zi}] = 0$, and that $E[u_i^2] = 1, E[e_{yi}^2] = \sigma_y^2,$ and $E[e_{zi}^2] = \sigma_z^2$. This gives us a variance matrix across the two equations of:

$$
\Sigma = \left[\begin{array}{cc}
\sigma_y^2 + \pi_y^2 & \pi_y\pi_z \\
\pi_y\pi_z & \sigma_z^2 + \pi_z^2
\end{array} \right]
$$

Note that I have assumed that the distribution of the latent variable $u_i$ is known to be a standard normal random variable. This is because the parameter $\pi$, which captures correlation across the equations, cannot be separately identified in the model. Anyways, given that we have observed. This idea borrows a bit from [Heckman et. al.](http://www.ncbi.nlm.nih.gov/pubmed/24187431) 

Given this latent variable, one way of writing down the full likelihood of our model is:

$$
LL =\Pi_{i=1}^N \int_u \frac{e^{-\frac{(y_i-\beta X_i-\delta z_i-\pi u_i)}{2\sigma_y^2}}}{\sqrt{2\pi\sigma_y^2}}
     \frac{e^{-\frac{(z_i-\eta W_i-u_i)}{2\sigma_z^2}}}{\sqrt{2\pi\sigma_z^2}} \frac{e^{-\frac{u_i^2}{2}}}{\sqrt{2\pi}}du_i  \times Prior(\beta,\pi,\delta,\sigma)
$$

Note that we have to integrate out the unobserved latent variable. In fact, this suggests that one way to estimate the model - just compute the above, and then use it as part of a likelihood routine. 

While there is nothing wrong with that, a Bayesian approach works a little differently, and is also easily extendable to other possibilities, dichotomous outcomes, different sorts of distributions, etc. So, let's now present an Gibbs sampling approach based on the above. The idea is to treat the latent variable itself as something that we want to draw and estimate along with everything else. 

## Sampling algorithm

1. Conditional on $X,y,u$ and $\sigma_y$, and a normal prior $ \beta,\delta = (\beta_p,\delta) \sim N(0,T_y) $ , $\beta$ and $\delta$ are just regular regression parameters, so they follow a multivariate normal distribution. That is, take $X_u = X,u$ and $Wu = W,u$ and let $\beta_d = \beta,\delta$. Then, we have the conditional distribution of $\beta_d$ as MVN, with mean $(X_u'X_u+\sigma_y^2T_y^{-1})^{-1}X_u'y$ and variance $(Xu'Xu + \sigma_y^2T^{-1})^{-1}\sigma_y$. 
2. Conditional on $W,u$, and $\sigma_z$, we see that $\eta$ also follows a normal with mean $(W'W+\sigma_zT_z^{-1})^{-1}W'(z-u)$ and variance $\sigma_z^2(W'W+\sigma_zT_z^{-1})^{-1}$. 
3. Conditional on everything else, and assuming the usual completely uninformative gamma prior gives us the distribution of $\sigma_y^2$ and $\sigma_z^2$ given everything else, as inverse gamma, with shape parameter $\frac{n}{2}$ and scale parameters $\alpha = \frac{n}{2}$, and $\beta = \frac{e_k'e_k}{2}$, where $e_k, k = y,z$ indicate error terms from our linear equations. 
4. Finally, given everything else, the $u_i$ can be shown to be normal with mean:
$$
\mu_i = \frac{(y_i - X_i\beta)\frac{\pi}{\sigma_y^2}+ \frac{z_i-W_i\eta}{\sigma_z^2}}{\frac{\pi^2}{\sigma_y^2} + \frac{1}{\sigma_z^2} + 1}
$$
and variance 
$$
\frac{1}{\frac{\pi^2}{\sigma_y^2} + \frac{1}{\sigma_z^2} + 1}
$$

So, the idea is pretty simple - just put some starting values in place and draw from the above distributions. 

## A Stata implementation

Here is an example of some `Stata` code that implements all of this in `Mata` [and using the awesome ipystata package](https://github.com/TiesdeKok/ipystata). We begin by just pulling in some data and taking a look at it. This is an example I just pulled right from the help file. The idea is to investigate the relationship between wages and education.

```python
import ipystata
```

```python
%%stata 
use http://www.stata.com/data/jwooldridge/eacsap/card.dta, clear
set more off
describe
```

Start with just a basic regression, trying to explain the log wage on a whole host of other things. 

```python
%%stata 
regress lwage educ expersq black south smsa reg661-reg668 smsa66, vce(r)
```

Let's add to this a basic IV regression. Here, we are going to use `nearc4` for an instrument for education. 

```python
%%stata 
ivregress 2sls lwage (educ=nearc2 nearc4) exper expersq black south smsa reg661-reg668, first
```

Fair enough...it looks as though education has a strong positive impact on wages. But are we convinced that there isn't some unobserved heterogeneity or correlation across equations? Let's see what our Bayesian method can tell us about this. 

We begin by reading the data into `Mata`, and adding a constant to each equation:

```python
%%stata --mata
    st_view(y=., ., "lwage")
    st_view(t=., ., "educ")
    st_view(X=., ., "educ exper expersq black south smsa reg661-reg668 smsa66")
    st_view(W=., ., "exper expersq black south smsa reg661-reg668 smsa66 nearc2 nearc4")
    X = X, J(rows(y), 1, 1)
    W = W, J(rows(y), 1, 1)
```

We will use the basic iv estimates as starting values, so:

```python
%%stata 
reg lwage educ exper expersq black south smsa reg661-reg668 smsa66
mat binit = e(b)
reg educ exper expersq black south smsa reg661-reg668 smsa66 nearc2 nearc4
mat einit = e(b)
```

Send those estimates to `Mata`:

```python
%%stata --mata
    b = st_matrix("binit"), 0
    e = st_matrix("einit"), 0

    nb = cols(b)
    ne = cols(e)
```

Set up a bunch of placeholders, some diffuse priors for parameters, and pick a random seed so results are reproducable. We also set the number of draws at 5000 and just draw a bunch of standard normal random variables for our initial hack at the unobserved terms. I just pick an initial value of one for the variances. 

```python
%%stata --mata
    bHold        = J(0, nb, .)
    eHold        = J(0, ne, .)
    syHold =       J(0, 1, .)
    stHold =       J(0, 1, .)
    piHold =       J(0, 2, .)

    Tb = I(nb)*1000
    Te = I(ne)*1000

    draws = 5000
    ui = rnormal(rows(y),1,0,1)
    sy2 = 1
    st2 = 1
```

## The Algorithm

As a practical matter, I draw first the values of $u_i$, then $b$, then $e$, and then the two variance terms from the inverse gamma distribution (actually, drawn from gamma with inverted parameters). 

```python
%%stata --mata
       bOnly =b[1..nb-1]
       eOnly =e[1..ne-1]
       piy = b[nb]
       pit = e[ne]
       resy = y-X*bOnly'
       rest = t-W*eOnly'
       A = piy^2/sy2 + pit^2/st2 + 1
       B = piy*resy/sy2 + pit*rest/st2
       mi = B:/A
       sd = 1/A
       ui = mi + rnormal(rows(mi),1,0,1)*sqrt(sd)
    
       Xu = (X, ui)
       mb = invsym(Xu'Xu + invsym(Tb))*Xu'y
       vb = sy2*invsym(Xu'Xu + invsym(Tb))
       cholb = cholesky(vb)
       b = mb + cholb*rnormal(nb, 1, 0, 1)
       b = b'
                                  
       Wu = (W, ui)
       me = invsym(Wu'Wu + invsym(Te))*Wu't
       ve = st2*invsym(Wu'Wu + invsym(Te))
       chole = cholesky(ve)
       e = me +chole*rnormal(ne, 1, 0, 1)
       e = e'

       shp = rows(y)/2
       scl = (y-Xu*b')'(y-Xu*b')/2
       scl = 1/scl
       sy2 = rgamma(1, 1, shp, scl)

       scl = (t - Wu*e')'(t - Wu*e')/2
       scl = 1/scl
       st2 = rgamma(1, 1, shp, scl)
                             
       bHold  =  bHold \ b
       eHold  =  eHold \ e
       syHold =  syHold \ sy2
       stHold =  stHold \ st2
       piHold =  piHold \ (b[nb], e[ne])
```

```python
%%stata --mata
   for (d=1;d<=draws;d++) {
       bOnly =b[1..nb-1]
       eOnly =e[1..ne-1]
       piy = b[nb]
       pit = e[ne]
       resy = y-X*bOnly'
       rest = t-W*eOnly'
       A = piy^2/sy2 + pit^2/st2 + 1
       B = piy*resy/sy2 + pit*rest/st2
       mi = B:/A
       sd = 1/A
       ui = mi + rnormal(rows(mi),1,0,1)*sqrt(sd)
    
       Xu = (X, ui)
       mb = invsym(Xu'Xu + invsym(Tb))*Xu'y
       vb = sy2*invsym(Xu'Xu + invsym(Tb))
       cholb = cholesky(vb)
       b = mb + cholb*rnormal(nb, 1, 0, 1)
       b = b'
                                  
       Wu = (W, ui)
       me = invsym(Wu'Wu + invsym(Te))*Wu't
       ve = st2*invsym(Wu'Wu + invsym(Te))
       chole = cholesky(ve)
       e = me +chole*rnormal(ne, 1, 0, 1)
       e = e'

       shp = rows(y)/2
       scl = (y-Xu*b')'(y-Xu*b')/2
       scl = 1/scl
       sy2 = rgamma(1, 1, shp, scl)

       scl = (t - Wu*e')'(t - Wu*e')/2
       scl = 1/scl
       st2 = rgamma(1, 1, shp, scl)
                             
       bHold  =  bHold \ b
       eHold  =  eHold \ e
       syHold =  syHold \ sy2
       stHold =  stHold \ st2
       piHold =  piHold \ (b[nb], e[ne])
    } 
```

We now have produced a set of draws. What can we do with them now? Well, the first and obvious thing to do is to take a look at the coefficient estimates. An easy way of doing this is to pull all of the variables into `Stata`:

```python
%%stata
preserve
clear
getmata (b*) = bHold
getmata (e*) = eHold
getmata (sy) = syHold
getmata (st) = stHold
```

Once all the variables are read in, we can summarize them and view these summary statistics as estimates of sorts. So:

```python
%%stata
sum *
```

There isn't a whole lot of interest here...but it is also neat to take a look at what the simulated distribution of the treatment effect parameter (`b1 above`) is. We get:

```python
%%stata --graph
hist b1
```

Once again, the above is the simulated distribution of the treatment effect parameter. Another thing we can do is take a look at what our random effects look like (the unobserved heterogeneity parameter). The last draw of this parameter is still held in `Mata`, so:

```python
%%stata
restore
preserve
clear
getmata ui
```

```python
%%stata --gr
hist ui
```

Let's also have a look at the estimated distribution of our $\pi$ parameter, which characterizes cross-equation correlation:

```python
%%stata --gr
restore
preserve
clear
getmata (piy pit) = piHold
hist piy
```

Another worthwhile thing to do is to check out how our parameters moves as the draws progress. We can do this using simple time-series plots:

```python
%%stata --gr
gen t = _n
tsset t
tsline piy pit
```

We see that our algorithm seems to converge to a situation in which our parameters have stable values quite quickly. We also see that the value for $piy$ is quite small, while that for $pit$ is significantly in the -1.6 or so range. What does this mean? Let's first think about our latent parameters and our variances. We have: 

```python
%%stata 
getmata sy = syHold
getmata st = stHold
sum sy 
scalar SY = r(mean)
sum st
scalar ST = r(mean)
sum piy 
scalar PIY = r(mean)
sum pit
scalar PIT = r(mean)
disp PIT
scalar v11 = SY + PIY^2
scalar v12 = PIY*PIT
scalar v22 = ST + PIT^2
disp v11
disp v12
disp v22
mat V = (v11, v12) \ (v12, v22)
mat list V
```

We see that there really isn't all that much cause for concern - it looks like the cross-equation correlation is more or less zero.  

It seems more or less normal, which is what we assumed it would look like. 

## Conclusions and next steps

This simple example is hopefully enough to get some of the basic ideas in place and give a feel for how a Bayesian estimation procedure might work. However, it is hard to see what the usefulness of the method is in this simple case, as estimation with correlation between equations is fairly simple in Maximum Likelihood. 

In the next few examples, it will become more obvious how the Bayesian method helps. There are two more notebooks which discuss cases with increasing complexity. So, we first discuss an example with a dichotomous treatment and continuous response (i.e., assignment to a control and treatment group), and then an example with a dichotomous treatment and response (i.e., treatment-control, good or bad outcome). In these cases, the use of latent variables allows us to distinguish between different types of cross-equation correlation, and also gives a little insight into things we never observe.

Also, we have not presented exact estimates for treatment effects or the distribution of these effects. We could probably get these as well, and even form confidence intervals for the individual $u_i$ terms. This might produce some really interesting distributional results.
